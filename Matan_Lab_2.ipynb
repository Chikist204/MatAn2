```python
from numpy import pi, cos, sin, arange, array
from matplotlib import pyplot as plt
```

```python
def draw(X, y, x1, x2):
    plt.plot(X, y, linewidth=3, color='b')
    plt.grid()
    y = 2 * x1
    plt.plot(x1, y, linewidth=3, color='r')
    y = x2 - x2 + 1
    plt.plot(x2, y, linewidth=3, color='r')
```

Дана функция: $$f(x)=\begin{cases}
\text{2x}, & x\in \left[0, 1\right) \\
\text{1}, & x\in \left[1, 2\right]	
\end{cases}.$$

## Общий тригонометрический ряд

Функцию можно представить в виде следующего тригонометрического ряда: $$f(x)=\frac{a_0}{2}+\sum_{i=1}^{\infty}a_k\cos(\frac{\pi kx}{l})+b_k\sin(\frac{\pi kx}{l}),$$ где $$a_0 = \frac{1}{l}\int_{-l}^{l}f(x)dx,$$
$$a_k = \frac{1}{l}\int_{-l}^{l}f(x)\cos(\frac{\pi kx}{l})dx,$$
$$b_k = \frac{1}{l}\int_{-l}^{l}f(x)\sin(\frac{\pi kx}{l})dx,$$
$$T = 2, l = 1$$

$$a_0 = \int_{0}^{2}f(x)dx = 2\int_{0}^{1}xdx + \int_{1}^{2}1dx=2$$
$$$$
$$a_k = \int_{0}^{2}f(x)\cos(\pi kx)dx=2\int_{0}^{1}x\cos(\pi kx)dx + \int_{1}^{2}\cos(\pi kx)dx=\left| \begin{array}{c} u = x \\ du = dx \\ dv = \cos(\pi kx)\ dx \\ v = \frac{1}{\pi k}\sin(\pi kx) \end{array}\right|=$$$$=2\frac{x}{\pi k}\sin(\pi kx)\Big|_0^1 - \frac{2}{\pi k} \int_{0}^{1}\sin(\pi kx)dx +\int_{1}^{2}\cos(\pi kx)dx = 0 + 2\frac{\cos(\pi k) - 1}{\pi^2k^2} + 0 = \underline{ 2\frac{(-1)^k - 1}{\pi^2k^2}}$$ $$$$ $$$$
$$b_k = \int_{0}^{2}f(x)\sin(\pi kx)dx=2\int_{0}^{1}x\sin(\pi kx)dx + \int_{1}^{2}\sin(\pi kx)dx=\left| \begin{array}{c} u = x \\ du = dx \\ dv = \sin(\pi kx)\ dx \\ v = \frac{-1}{\pi k}\cos(\pi kx) \end{array}\right|=$$$$=-2\frac{x}{\pi k}\cos(\pi kx)\Big|_0^1 - \frac{2}{\pi k} \int_{0}^{1}\cos(\pi kx)dx +\int_{1}^{2}\sin(\pi kx)dx = \frac{-2\cos(\pi k)}{\pi k} + 0 + \frac{\cos(\pi k) - cos(2\pi k)}{\pi k} = \underline{ -\frac{(-1)^k + 1}{\pi k}}$$

Тогда Тригонометрический ряд будет иметь вид: $$f(x)=1 + \sum_{1}^{\infty}2\frac{(-1)^k - 1}{\pi^2k^2} \cos(\pi kx) - \frac{(-1)^k + 1}{\pi k} \sin(\pi kx)$$

Сумма ряда Фурье: $$S(x)=\begin{cases}
\frac{f(x-0)+f(x+0)}{2}=1.5, & x = 1 \\
1 + \sum_{1}^{\infty}2\frac{(-1)^k - 1}{\pi^2k^2} \cos(\pi kx) - \frac{(-1)^k + 1}{\pi k} \sin(\pi kx), & x\neq1 
\end{cases}.$$

### График функции и частичных сумм ($S_5, S_{10}, S_{50}$)

```python
a0 = 2
s = 0
X = arange(0, 2, 0.1)
y = []
for x in X:
    f = a0 / 2
    s = 0
    for k in range(1, 5):
        s += (2 * ((-1) ** k - 1) / (pi**2 * k**2)) * cos(k * pi * x) + ((-((- 1) ** k) - 1) / (pi * k)) * sin(pi * k * x)
    f += s
    y.append(f)
y = array(y)
x1 = arange(0, 1, 0.01)
x2 = arange(1, 2.1, 0.5)
draw(X, y, x1, x2)
plt.show()
```

## Ряд Фурье по синусам

Представим функцию в виде ряда Фурье по синусам: $$f(x)=\sum_{i=1}^{\infty}b_k\sin(\frac{\pi kx}{l}),$$ где 
$$b_k = \frac{2}{l}\int_{-l}^{l}f(x)\sin(\frac{\pi kx}{l})dx,$$
$$T = 4, l = 2$$

$$b_k = \int_{0}^{2}f(x)\sin(\frac{\pi kx}{2})dx=2\int_{0}^{1}x\sin(\frac{\pi kx}{2})dx + \int_{1}^{2}\sin(\frac{\pi kx}{2})dx=\left| \begin{array}{c} u = x \\ du = dx \\ dv = \sin(\frac{\pi kx}{2})\ dx \\ v = \frac{-2}{\pi k}\cos(\frac{\pi kx}{2}) \end{array}\right|=$$$$=\frac{-4x}{\pi k}\cos(\frac{\pi kx}{2})\Big|_0^1 - \frac{4}{\pi k} \int_{0}^{1}\cos(\frac{\pi kx}{2})dx +\int_{1}^{2}\sin(\frac{\pi kx}{2})dx = \frac{-4\pi k\cos(\frac{\pi k}{2}) + 8\sin(\frac{\pi k}{2})}{\pi^2 k^2} + \frac{2\cos(\frac{\pi k}{2}) - 2\cos(\pi k)}{\pi k} = \frac{8\sin(\frac{\pi k}{2})-2\pi k\cos(\frac{\pi k}{2})-2\pi k\cos(\pi k)}{\pi^2 k^2}$$

Тогда ряд по синусам: $$f(x)=\sum_{1}^{\infty}\frac{8\sin(\frac{\pi k}{2})-2\pi k\cos(\frac{\pi k}{2})-2\pi k\cos(\pi k)}{\pi^2 k^2} \sin(\frac{\pi kx}{2})$$

Сумма ряда Фурье: $$S(x)=\begin{cases}
1.5, & x = 1 \\
-1.5, & x = -1 \\
\sum_{1}^{\infty}\frac{8\sin(\frac{\pi k}{2})-2\pi k\cos(\frac{\pi k}{2})-2\pi k\cos(\pi k)}{\pi^2 k^2} \sin(\frac{\pi kx}{2}), & x\neq1 
\end{cases}.$$

### График функции и частичных сумм ($S_5, S_{10}, S_{50}$)

```python
s = 0
X = arange(-2, 2.1, 0.1)
y = []
for x in X:
    f = 0
    s = 0
    for k in range(1, 6):
        if k % 4 == 0:
            s += ((-2 * pi * k * 2) / (pi**2 * k**2)) * sin(pi * k * x / 2)
        elif k % 2 == 0:
            s += ((-2 * pi * k * 0) / (pi**2 * k**2)) * sin(pi * k * x / 2)
        else:
            s += ((-2 * pi * k * ((-1) ** k + cos(pi * k / 2)) + 8 * sin(pi * k / 2)) / (pi**2 * k**2)) * sin(pi * k / 2 * x)
    f += s
    y.append(f)
y = array(y)
x1 = arange(-1, 1, 0.01)
x2 = arange(1, 2.1, 0.5)
draw(X, y, x1, x2)
x2_2 = arange(-2, -0.9, 0.5)
y = x2_2 - x2_2 - 1
plt.plot(x2_2, y, linewidth=3, color='r')
plt.show()
```

## Ряд Фурье по косинусам

Представим функцию в виде ряда Фурье по косинусам: $$f(x)=\frac{a_0}{2} + \sum_{i=1}^{\infty}a_k\cos(\frac{\pi kx}{l}),$$ где 
$$a_k = \frac{2}{l}\int_{-l}^{l}f(x)\cos(\frac{\pi kx}{l})dx,$$
$$T = 4, l = 2$$

$$a_k = \int_{0}^{2}f(x)\cos(\frac{\pi kx}{2})dx=2\int_{0}^{1}x\cos(\frac{\pi kx}{2})dx + \int_{1}^{2}\cos(\frac{\pi kx}{2})dx=$$ $$=\frac{4\pi k\sin(\frac{\pi k}{2}) + 8\cos(\frac{\pi k}{2}) - 8}{\pi^2 k^2}+\frac{2\sin(\pi k) - 2\sin(\frac{\pi k}{2})}{\pi k}=\frac{2\pi k\sin(\frac{\pi k}{2}) + 8\cos(\frac{\pi k}{2}) - 8}{\pi^2 k^2}$$

Тогда ряд по синусам: $$f(x)=1 + \sum_{1}^{\infty}\frac{2\pi k\sin(\frac{\pi k}{2}) + 8\cos(\frac{\pi k}{2}) - 8}{\pi^2 k^2} \cos(\frac{\pi kx}{2})$$

Сумма ряда Фурье: $$S(x)=\begin{cases}
1.5, & x = \pm 1 \\
1 + \sum_{1}^{\infty}\frac{2\pi k\sin(\frac{\pi k}{2}) + 8\cos(\frac{\pi k}{2}) - 8}{\pi^2 k^2} \cos(\frac{\pi kx}{2}), & x\neq \pm1 
\end{cases}.$$

### График функции и частичных сумм ($S_5, S_{10}, S_{50}$)

```python
a0 = 2
X = arange(-2, 2.1, 0.1)
y = []
for x in X:
    f = a0 / 2
    s = 0
    for k in range(1, 6):
        if k % 4 == 0:
            s += 0
        if k % 2 == 0:
            s += (8 * (-2) / (pi**2 * k**2)) * cos(pi * k / 2 * x)
        else:
            s += ((8 * (-1) + 2 * pi * k * sin(pi * k / 2)) / (pi**2 * k**2)) * cos(pi * k / 2 * x)
    f += s
    y.append(f)
y = array(y)
x1 = arange(0, 1, 0.01)
x2 = arange(1, 2.1, 0.5)
draw(X, y, x1, x2)
x1_2 = arange(-1, 0.1, 0.5)
y = 2 * abs(x1_2)
plt.plot(x1_2, y, linewidth=3, color='r')
x2_2 = arange(-2, -0.9, 0.5)
y = x2_2 - x2_2 + 1
plt.plot(x2_2, y, linewidth=3, color='r')
plt.show()
```
